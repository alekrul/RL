```
class randomPolicy:
  def __init__(self, N):
    self.N = N
  
  def act(self, s):
    self.N += 1
    return np.random.randint(0,4)
```

```
class betterPolicy:
  def __init__(self, N):
    self.N = N
  def act(self, s):
    if s == 36:
      return 0
    elif s == 11 or s == 23 or s == 35:
      return 2
    elif (s >= 37 and s<= 46):
      return 0
    else:
      return 1
```

```
env = gym.make('CliffWalking-v0')
s = env.reset()
N = 0
policy = betterPolicy(N)
done = False
EPISODE_LENGTH = 100
cumulative_reward = 0
for i in range(EPISODE_LENGTH):
  action = policy.act(s)
  print("-------------------------------------------")
  print("state: ", s)
  print("Action: ", action)
  print("Iteration: ", policy.N)
  s, r, done, _ = env.step(action)
  cumulative_reward += r
  print("rewards: ", r)
  print("cumulative reward: ", cumulative_reward)
  print("New state:", s)
  env.render()
  if done:
    break
```
